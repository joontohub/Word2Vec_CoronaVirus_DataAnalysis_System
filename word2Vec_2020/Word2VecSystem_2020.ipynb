```python
import pandas as pd

pd.set_option('max_colwidth',500)

# 파일을 DataFrame으로 전환
flu_rawdata = pd.read_csv('moviere_flu_2020.csv', header=0,encoding='utf-8', dtype=str, engine='python')
contagion_rawdata = pd.read_csv('moviere_contagion_2020.csv', header=0,encoding='utf-8', dtype=str, engine='python')
```

```python
## 공백값(NAN) 제거
flu_rawtext = flu_rawtext.dropna(axis=0)
```

```python
contagion_rawtext = contagion_rawtext.dropna(axis=0)
```

```python
import time
start_vect=time.time()

from konlpy.tag import Okt
from konlpy.utils import pprint
okt = Okt()

pos = lambda d: [''.join(p) for p in okt.nouns(d)]
flu_texts_ko = [pos(result) for result in flu_rawtext['text']]

contagion_texts_ko = [pos(result) for result in contagion_rawtext['text']]

print("training Runtime: %0.2f Minutes"%((time.time() - start_vect)/60))
```

```python
flu_texts_ko
```

```python
contagion_texts_ko
```

```python
start_vect=time.time()

# training 모델 훈련

from gensim.models import word2vec
flu_wv_model_ko = word2vec.Word2Vec(flu_texts_ko, size=1000, window=10, min_count=0, iter=10, sg=1)

contagion_wv_model_ko = word2vec.Word2Vec(contagion_texts_ko, size=1000, window=10, min_count=0, iter=10, sg=1)
## size=차원의 숫자(올라갈수록 상세히 하지만 느림) 100
## window=주변을 보는 범위(앞뒤 5) 
## min_count=최소 출현 빈도 100
## workers=사용하는 CPU 코어 숫자
## iter = 반복횟수 10
## sg= CBOW=0 / Skip-Gram=1

#from gensim.models import FastText
#wv_model_ko = FastText(texts_ko, size=1000, window=10, min_count=0, iter=10, sg=1)


flu_wv_model_ko.init_sims(replace=True)
flu_wv_model_ko.save('flu_ko_word2vec_e.model')

contagion_wv_model_ko.init_sims(replace=True)
contagion_wv_model_ko.save('contagion_ko_word2vec_e.model')

print("training Runtime: %0.2f Minutes"%((time.time() - start_vect)/60))
```

```python


from sklearn.manifold import TSNE
import matplotlib as mpl
import matplotlib.pyplot as plt
import gensim 
import gensim.models as g

# 그래프에서 마이너스 폰트 깨지는 문제에 대한 대처
mpl.rcParams['axes.unicode_minus'] = False

flu_model_name = 'flu_ko_word2vec_e.model'
contagion_model_name = 'contagion_ko_word2vec_e.model'
flu_model = g.Doc2Vec.load(flu_model_name)
contagion_model = g.Doc2Vec.load(contagion_model_name)

flu_vocab = list(flu_model.wv.vocab)
contagion_vocab = list(contagion_model.wv.vocab)
flu_X = flu_model[flu_vocab]
contagion_X = contagion_model[contagion_vocab]

```

```python
from sklearn.manifold import TSNE
import matplotlib as mpl

import gensim 
import gensim.models as g

# 그래프에서 마이너스 폰트 깨지는 문제에 대한 대처
mpl.rcParams['axes.unicode_minus'] = False

flu_model_name = 'flu_ko_word2vec_e.model'

flu_model = g.Doc2Vec.load(flu_model_name)

flu_vocab = list(flu_model.wv.vocab)
flu_X = flu_model[flu_vocab]
```

```python
mpl.rcParams['axes.unicode_minus'] = False

contagion_model_name = 'contagion_ko_word2vec_e.model'

contagion_model = g.Doc2Vec.load(contagion_model_name)

contagion_vocab = list(contagion_model.wv.vocab)
contagion_X = contagion_model[contagion_vocab]
```

```python
flu_tsne = TSNE(n_components=3)
contagion_tsne = TSNE(n_components=3)
```

```python
import time
start_vect=time.time()

flu_X_tsne = flu_tsne.fit_transform(flu_X)
# flu_X_tsne = flu_tsne.fit_transform(X)
contagion_X_tsne = contagion_tsne.fit_transform(contagion_X)
print("training Runtime: %0.2f Minutes"%((time.time() - start_vect)/60))
```

```python
flu_df = pd.DataFrame(flu_X_tsne, index=flu_vocab, columns=['x', 'y', 'z'])
flu_df.shape
```
